```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:, [2,3]]
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1, stratify=y)
```

```python
tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=1)
tree.fit(X_train, y_train)
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions.plot_decision_regions(X_combined, y_combined, classifier=tree, test_idx=range(105, 150))
plt.xlabel('Dugo patka [cm]')
plt.ylabel('Szeroko patka [cm]')
plt.legend(loc='upper left')
plt.show()
```

```python
# wizualizacja drzewa za pomoc biblioteki graphviz
```

```python
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
dot_data = export_graphviz(tree, 
                           filled=True, 
                           rounded=True,
                           class_names=['Setosa', 'Versiocolor', 'Virginica'],
                           feature_names=['Dugo patka', 'Szeroko patka'], out_file=None)
graph = graph_from_dot_data(dot_data)
graph.write_png('drzewo.png')


```
